```python
x = [1,2,3]
y = [1,5,20]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.title('Sales Indication')
plt.xlabel('Sales per Year')
plt.ylabel('Revenue per Year')
plt.plot(x,y)
plt.show()
```
